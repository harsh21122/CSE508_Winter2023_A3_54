```python
from tabulate import tabulate
import networkx as nx
import numpy as np
import pandas as pd
```

```python
G = nx.DiGraph()
```

```python
with open('p2p-Gnutella08.txt', 'r') as file:
    for idx, line in enumerate(file):
        if idx > 3:
            ll = line.strip().split()
#             print(ll[0], ll[1])
            G.add_edge(ll[0], ll[1])
```

```python
print("Number of nodes: ", G.number_of_nodes())
print("Number of edges: ", G.number_of_edges())
print("Is Directed : ", G.is_directed())
```

```python
pagerank = nx.pagerank(G)
```

```python
hits = nx.hits(G)
```

```python
df = pd.DataFrame(columns = ['Node', 'PageRank', 'Authority', 'Hub'])
for idx in range(G.number_of_nodes()):
    node = str(idx)
    pg = pagerank[node]
    aut = hits[0][node]
    hub = hits[1][node]
    df = df.append({'Node': node, 'PageRank': pg, 'Authority': aut, 'Hub': hub}, ignore_index=True)
```

```python
print(tabulate(df, headers='keys', tablefmt='psql', floatfmt='.5f'))
```

```python
 # If a node has a higher PageRank score than HITS score,
    # it indicates that the node is well-connected and has many quality links pointing to it
    
# Good autority means many hubs point to it i.e. many incoming link
# Good hub points towards many good authority. So how many outgoing links?
# HITS score tells wheter good authority or hub?
```

```python
import snap

# load the graph from the file
filename = "p2p-Gnutella08.txt"
graph = snap.LoadEdgeList(snap.PUNGraph, filename, 0, 1)

# calculate the PageRank scores for the nodes in the graph
PRankH = snap.TIntFltH()
snap.GetPageRank(graph, PRankH)

# print the PageRank scores for the first 10 nodes
for node in PRankH:
    print("Node ID: %d, PageRank Score: %f" % (node, PRankH[node]))
    if node >= 10:
        break
        
NIdHubH = snap.TIntFltH()
NIdAuthH = snap.TIntFltH()
snap.GetHits(graph, NIdHubH, NIdAuthH)

# print the HITS scores for the first 10 nodes
for node in NIdAuthH:
    print("Node ID: %d, HITS Score: %f %f" % (node, NIdAuthH[node], NIdHubH[node]))
    if node >= 10:
        break
```
